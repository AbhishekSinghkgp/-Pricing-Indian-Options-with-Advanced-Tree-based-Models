```python
df = df.rename(columns=lambda x: x.strip())
df = df.dropna(subset=['Date','Expiry','Strike Price','Close','Underlying Value'])
df = df[(df['Underlying Value'] != '-') & 
        (df['Date'] != '-') & 
        (df['Expiry'] != '-') & 
        (df['Strike Price'] != '-') & 
        (df['Close'] != '-')]
```

```python
# time_to_maturity
df['Date'] = pd.to_datetime(df['Date'])
df['Expiry'] = pd.to_datetime(df['Expiry'])

df['time_to_maturity_days'] = (df['Expiry'] - df['Date']).dt.days
```

```python
# option type
encoding_map = {'CE': 1, 'PE': 0}
df['Option type'] = df['Option type'].map(encoding_map)
df.rename(columns = {'Option type' : 'Option_type'}, inplace = True)
```

```python
df['time_to_maturity_years'] = df['time_to_maturity_days'] / 365
```

```python
# Function to interpolate the risk-free rate
def interpolate_risk_free_rate(row, risk_f_df,z):
    # Find the closest lower and upper years
    lower_year = int(np.floor(row['time_to_maturity_years']))
    upper_year = int(np.ceil(row['time_to_maturity_years']))

    # If the maturity is less than 1 year, use the 1-year rate
    if lower_year == 0:
        return risk_f_df.loc[1, row['Date'].month].iloc[z]
    
    # If exact year matches, return the rate
    if lower_year == row['time_to_maturity_years']:
        return risk_f_df.loc[lower_year, row['Date'].month].iloc[z]
    
    
    # Interpolate between the two years
    lower_rate = risk_f_df.loc[lower_year, row['Date'].month].iloc[z]
    upper_rate = risk_f_df.loc[upper_year, row['Date'].month].iloc[z]
    
    # If either rate is NaN, we cannot interpolate; return NaN
    if pd.isna(lower_rate) or pd.isna(upper_rate):
        return np.nan

    # Calculate the weight for interpolation based on the distance to the lower and upper years
    weight = (row['time_to_maturity_years'] - lower_year) / (upper_year - lower_year)

    # Return the interpolated rate
    return (1 - weight) * lower_rate + weight * upper_rate

```

```python
# Apply the function to each row in 'df' to create the new column
df['riskfree_rates'] = df.apply(lambda row: interpolate_risk_free_rate(row, risk_f,4), axis=1)
```
